```julia
using Plots
```

```julia
include("../src/kf.jl")
include("../src/ngpkf.jl")
include("../src/Clarity.jl")
```

```julia
w_hat = NGPKF.initialize(Clarity.ngpkf_grid)
q_map = NGPKF.clarity_map(Clarity.ngpkf_grid, w_hat)
```

```julia
heatmap(Clarity.ngpkf_grid.xs, Clarity.ngpkf_grid.ys, q_map, title="Initial clarity map", xlabel="x", ylabel="y", clims=(0,1))
```

```julia
new_measurement_pos = [1.0, 4.0];
new_measurement_w = 3.4;

w_hat = NGPKF.predict(Clarity.ngpkf_grid, w_hat)
w_hat_new = NGPKF.correct(Clarity.ngpkf_grid, w_hat, new_measurement_pos, new_measurement_w)
q_map_new = NGPKF.clarity_map(Clarity.ngpkf_grid, w_hat_new)
```

```julia
heatmap(Clarity.ngpkf_grid.xs, Clarity.ngpkf_grid.ys, q_map_new, title="Initial clarity map", xlabel="x", ylabel="y", clims=(0,1))
```
